```python
%pip install matplotlib

import pandas as pd
import matplotlib.pyplot as plt

df = companies_data = pd.read_csv("../../data/article_company_breach/joined_articles_company_breaches_no_duplicates.csv")

# Group by the specified columns
grouped = df.groupby(["organisation", "records lost", "year", "breach_date", "total employee estimate"])

# Count the size of each group (number of articles)
group_sizes = grouped.size().reset_index(name="number of articles")

# Sort by 'records lost' for better visualization
group_sizes_sorted = group_sizes.sort_values("records lost")

# Plot the scatter plot (dots only, transparent)
plt.figure(figsize=(10, 6))
plt.scatter(
    group_sizes_sorted["records lost"],
    group_sizes_sorted["number of articles"],
    color="blue",
    alpha=0.6,
    s=100,
    label="Number of Articles"
)

# Add labels to each dot
for i in range(len(group_sizes_sorted)):
    plt.text(
        group_sizes_sorted["records lost"].iloc[i],  # X-coordinate
        group_sizes_sorted["number of articles"].iloc[i],  # Y-coordinate
        f"{group_sizes_sorted['organisation'].iloc[i]}, {group_sizes_sorted['year'].iloc[i]}",  # Label (organisation name)
        fontsize=9,
        ha="right",  # Align text to the right of the dot
        alpha=0.7
    )


plt.xscale("log")

# Add labels and title
plt.title("Records Lost vs Number of Articles", fontsize=16)
plt.xlabel("Records Lost", fontsize=12)
plt.ylabel("Number of Articles", fontsize=12)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.grid(axis="both", linestyle="--", alpha=0.7)
plt.legend(fontsize=12)

# Show the plot
plt.tight_layout()
plt.show()
```
